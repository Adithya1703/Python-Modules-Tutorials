```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create 1D arrays for x and y using logspace (logarithmic scale)
x = np.logspace(0, 2, 30, base=10)  # 10^0 to 10^2 => 1 to 100
y = np.logspace(0, 2, 30, base=10)

# Create 2D meshgrid for plotting
X, Y = np.meshgrid(x, y)

# Define a 2D function to evaluate over the grid
Z = np.log10(X**2 + Y**2)  # Just an example function to visualize

# Plotting a 3D surface plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_title('Surface Plot Using logspace() and meshgrid()')
ax.set_xlabel('X (log-spaced)')
ax.set_ylabel('Y (log-spaced)')
ax.set_zlabel('Z = log10(X + Y)')
fig.colorbar(surf, ax=ax, shrink=0.6, aspect=10)
plt.tight_layout()
plt.show()
```
